# Load Modules and Prepare Data

```python
%load_ext autoreload
%autoreload 2
from plot_utils import PyseroLoader
from plot_utils import SerumDilutionPlotter


data_dir = "../data/Results_5-20-2020/"
scienion_xlsx = "2020-05-18-17-59-01-COVID_May18_JVassay_analysis.xlsx"

# loads metadata and optical density data
psl = PyseroLoader(data_dir, scienion_xlsx)
psl.LoadMeta()
psl.LoadOD()

# merges metadata and od data
pysero_frame = psl.FullPyseroOutput()
scienion_frame = psl.FullScienionOutput()
```

# Fit Curves for Plotting

```python
sdp = SerumDilutionPlotter()
frame_pysero = sdp.fit(pysero_frame)
frame_scienion = sdp.fit(scienion_frame)
```

# Plot Antigens by giving a secondary dilution + a section of its name

## Pysero RBD

```python
fig_pysero_rbd_00005 = sdp.plot_antigen(frame_pysero, 0.00005, 'RBD')

fig_pysero_rbd_00005
```

## Scienion all SARS

```python
# Note that the antigen argument will match all antigens that contain the string
## This can be used to capture antigens at various dilutions
fig_scienion_SARS_00005 = sdp.plot_antigen(frame_scienion, 0.00005, 'SARS')

fig_scienion_SARS_00005
```

# Figures can be written to HTML or PDF

```python
# html
sdp.write_plot(fig_pysero_rbd_00005, "../figures/pysero_rbd_00005.html", format = 'html')

# pdf
sdp.write_plot(fig_pysero_rbd_00005, "../figures/pysero_rbd_00005.pdf", format = 'pdf')
```
